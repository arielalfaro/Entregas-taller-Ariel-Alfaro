3. (15 puntos)Realiza un test de hipotesis para evaluar si las dos variables que se agregaron son
significativas para el modelo simultaneamente. Use significancia del 5%.

```r
# Variables agregadas: sales_mayor_promedio y sales_mayor_promedio_roe
# H0: beta1 = beta2 = 0
# Modelo restringido (sin las dos variables)
modelo_r <- lm(salary ~ sales + roe + ros, data = ceosal1)

# Modelo no restringido (con las dos variables)
modelo_ur <- lm(salary ~ sales + roe + ros + sales_mayor_promedio + sales_mayor_promedio_roe, data = ceosal1)

# Extraer R²
R2_r <- summary(modelo_r)$r.squared
R2_ur <- summary(modelo_ur)$r.squared

# Datos para el test F
n <- nrow(ceosal1)
q <- 2
k_ur <- 5
# Estadístico F
F_est <- ((R2_ur - R2_r) / q) / ((1 - R2_ur) / (n - k_ur - 1))

# Valor crítico F al 5%
F_critico <- qf(0.95, q, n - k_ur - 1)

#Resultados
cat("=== 2.3 Test conjunto ===\n")
cat("R² restringido:", round(R2_r, 5), "\n")
cat("R² no restringido:", round(R2_ur, 5), "\n")
cat("Estadístico F calculado:", round(F_est, 4), "\n")
cat("Valor crítico F (5%):", round(F_critico, 4), "\n")

if (F_est > F_critico) {
  cat("Conclusión: Se rechaza H0 → las variables son conjuntamente significativas\n")
} else {
  cat("Conclusión: No se rechaza H0 → las variables no son conjuntamente significativas\n")
}
```

4. (15 puntos) Volviendo a la regresión del salario en función de sales, roe y ros. Considere
dos grupos, Grupo 1: Aquellos clientes que sales es mayor al promedio, y Grupo 2: Aquellos
clientes que sales es menor al promedio. Realice un test de hipotesis para grupos al 5% de
significancia. Use el metodo que más le acomode.

```r
# Método: modelo con interacciones completas
# H0: los grupos son iguales (pendientes e intercepto)
# Crear variables de interacción
ceosal1$sales_D <- ceosal1$sales * ceosal1$sales_mayor_promedio
ceosal1$roe_D   <- ceosal1$roe   * ceosal1$sales_mayor_promedio
ceosal1$ros_D   <- ceosal1$ros   * ceosal1$sales_mayor_promedio

# Modelo no restringido (incluye dummy y todas las interacciones)
modelo_grupos_ur <- lm(salary ~ sales + roe + ros + sales_mayor_promedio + sales_D + roe_D + ros_D, data = ceosal1)

# Modelo restringido (sin diferencias entre grupos)
modelo_grupos_r <- lm(salary ~ sales + roe + ros, data = ceosal1)
# R²
R2_r_grupos <- summary(modelo_grupos_r)$r.squared
R2_ur_grupos <- summary(modelo_grupos_ur)$r.squared

# Datos para el test F
n_grupos <- nrow(ceosal1)
q_grupos <- 4
k_ur_grupos <- 7

# Estadístico F
F_est_grupos <- ((R2_ur_grupos - R2_r_grupos) / q_grupos) / ((1 - R2_ur_grupos) / (n_grupos - k_ur_grupos - 1))

# Valor crítico F al 5%
F_critico_grupos <- qf(0.95, q_grupos, n_grupos - k_ur_grupos - 1)

# Resultados
cat("\n=== 2.4 Test de grupos (sales mayor vs menor al promedio) ===\n")
cat("R² restringido:", round(R2_r_grupos, 5), "\n")
cat("R² no restringido:", round(R2_ur_grupos, 5), "\n")
cat("Estadístico F calculado:", round(F_est_grupos, 4), "\n")
cat("Valor crítico F (5%):", round(F_critico_grupos, 4), "\n")

if (F_est_grupos > F_critico_grupos) {
  cat("Conclusión: Se rechaza H0 → los grupos son diferentes\n")
} else {
  cat("Conclusión: No se rechaza H0 → los grupos son iguales\n")
}
```

5. (15 puntos) Usando la regresión del salario en función de sales, roe y ros, se distinguen los
grupos mediante las dummies; indus, finance, consprod y utility, que corresponden al tipo de
empresa. Realice mediante test de Chow, con 5% de significancia, si los modelos para cada
grupo son iguales.

```r
# Grupos: indus, finance, consprod, utility
# H0: todos los grupos tienen el mismo modelo
modelo_chow_r <- lm(salary ~ sales + roe + ros, data = ceosal1)

# Subconjuntos por grupo
indus_sub   <- subset(ceosal1, indus == 1)
finance_sub <- subset(ceosal1, finance == 1)
conprod_sub <- subset(ceosal1, consprod == 1)
utility_sub <- subset(ceosal1, utility == 1)
# Modelos por separado
modelo_indus   <- lm(salary ~ sales + roe + ros, data = indus_sub)
modelo_finance <- lm(salary ~ sales + roe + ros, data = finance_sub)
modelo_conprod <- lm(salary ~ sales + roe + ros, data = conprod_sub)
modelo_utility <- lm(salary ~ sales + roe + ros, data = utility_sub)
# Suma de cuadrados de residuos
SSR_r <- sum(residuals(modelo_chow_r)^2)
SSR_ur <- sum(residuals(modelo_indus)^2) +
         sum(residuals(modelo_finance)^2) +
         sum(residuals(modelo_conprod)^2) +
         sum(residuals(modelo_utility)^2)

# Grados de libertad
G <- 4
k <- 3
gl_num <- (G - 1) * (k + 1)
gl_den <- nrow(ceosal1) - G * (k + 1)

# Estadístico F
F_est_chow <- ((SSR_r - SSR_ur) / gl_num) / (SSR_ur / gl_den)
# Valor crítico F al 5%
F_critico_chow <- qf(0.95, gl_num, gl_den)

# Resultados
cat("\n=== 2.5 Test de Chow por tipo de empresa ===\n")
cat("SSR restringido (modelo conjunto):", round(SSR_r, 2), "\n")
cat("SSR no restringido (suma de grupos):", round(SSR_ur, 2), "\n")
cat("Estadístico F calculado:", round(F_est_chow, 4), "\n")
cat("Valor crítico F (5%):", round(F_critico_chow, 4), "\n")
cat("Grados de libertad:", gl_num, "y", gl_den, "\n")

if (F_est_chow > F_critico_chow) {
  cat("Conclusión: Se rechaza H0 → los modelos difieren por tipo de empresa\n")
} else {
  cat("Conclusión: No se rechaza H0 → los modelos son iguales entre tipos de empresa\n")
}
```
